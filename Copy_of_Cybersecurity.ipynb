```python
from sklearn.model_selection import train_test_split
import pandas as pd

df = pd.read_csv("/content/file_pe_headers.csv", on_bad_lines='skip') # Using an existing file
df.dropna(inplace=True) # Drop rows with any NaN values
y = df['Malware'] # Correct target column for file_pe_headers.csv
X = df.drop(['Name','Malware'], axis=1) # Correct feature columns for file_pe_headers.csv
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.2, random_state=31)
```
